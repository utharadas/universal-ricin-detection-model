```python
for i in range(len(bins2)-2):
    df_copy = results.copy()
    rg_low = bins2[i]
    rg_high = bins2[i+1]
    df_copy = df_copy[df_copy['Freq(Hz)']>rg_low]
    df_copy = df_copy[df_copy['Freq(Hz)']<rg_high]
    print (df_copy['correct'].mean())
    print (f"{rg_low} - {rg_high}")
```

```python
import matplotlib.pyplot as plt
```

```python
accuracy_by_bin = results.groupby('frequency_bin')['correct'].mean()

# Plot accuracy versus frequency range
plt.figure(figsize=(10, 6))
accuracy_by_bin.plot(kind='bar', color='skyblue')
plt.xlabel('Frequency Range')
plt.ylabel('Accuracy')
plt.title('Model Accuracy by Frequency Range')
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.show()
```
